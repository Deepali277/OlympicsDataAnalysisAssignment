# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
df.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
df = pd.read_csv("C:/Users/User/Downloads/summer.csv")
unique_city=len(df["City"].unique())
print("\nSo far Summer Olympics are held in",unique_city,"cities.\n","\nThey are:",df["City"].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data2=[]

for sport in df["Sport"].unique():
    data2.append([sport,len(df[(df.Sport==sport)&(df.Medal=="Gold")])])

d=pd.DataFrame(data2,columns=["Sport","No of gold medals"]).sort_values(by="No of gold medals",
ascending=False).head()

print("The top 5 Sport having most number of Gold Medals are:")       
for i in range(0,5):
    print("Rank",i+1,":",d["Sport"].iloc[i],"with",d["No of gold medals"].iloc[i],"Gold medals")

data2=d.plot(x="Sport",y="No of gold medals",kind="bar",figsize=(13,7),color="gold",edgecolor="k")
plt.xlabel("Sport",color="blue",fontsize=16)
plt.ylabel("Number of gold medals",color="blue",fontsize=16)
plt.title("Sport V/S Number of Gold Medals",color="green",fontweight='bold',fontsize=20)
plt.legend(fontsize=14)
```

### 3. Which sport is having most number of medals so far? (Top 5)
